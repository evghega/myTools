```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from sklearn.metrics.pairwise import cosine_similarity
```

```python

ds = pd.read_csv("./content-sample-data.csv")
print("The Dim of the data is: ", (ds.shape)) 
```

```python
ds.head
```

```python

tf = TfidfVectorizer(analyzer='word', ngram_range=(1, 3), min_df=0, stop_words='english')
tfidf_matrix = tf.fit_transform(ds['description'])

```

```python
print("The Dim of tfidf_matrix is: ", (tfidf_matrix.shape)) 
print(tfidf_matrix)

print (tf.vocabulary_)
```

```python

cosine_similarities = linear_kernel(tfidf_matrix)

print("The Dim of cosine_similarities is: ", (cosine_similarities.shape)) 
print(cosine_similarities)


results = {}

for idx, row in ds.iterrows():
    similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
    similar_items = [(cosine_similarities[idx][i], ds['id'][i]) for i in similar_indices]

    results[row['id']] = similar_items[1:]
    
print('done!')
```

```python

print("The Dim of results is: ", len(results)) 
print(results)
```

```python
cosine_similarities2 = cosine_similarity(tfidf_matrix)

results1 = {}

for idx, row in ds.iterrows():
    similar_indices = cosine_similarities2[idx].argsort()[:-100:-1]
    similar_items = [(cosine_similarities2[idx][i], ds['id'][i]) for i in similar_indices]

    results1[row['id']] = similar_items[1:]
    
print('done!')
```

```python

def item(id):
    return ds.loc[ds['id'] == id]['description'].tolist()[0].split(' - ')[0]

# Just reads the results out of the dictionary.
def recommend(item_id, num):
    print("Recommending " + str(num) + " products similar to " + item(item_id) + "...")
    print("-------")
    recs = results[item_id][:num]
    for rec in recs:
        print("Recommended: " + item(rec[1]) + " (score:" + str(rec[0]) + ")")

```

```python
recommend(item_id=11, num=5)
```
