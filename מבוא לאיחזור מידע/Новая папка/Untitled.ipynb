```python

from spellchecker import SpellChecker
spell = SpellChecker(distance=1)

val = spell.unknown(['Hello', 'there', 'are', 'u', 'here', 'wtf','schol'])
for word in val:
    print(spell.correction(word))
    
    
print("--------------------------------") 
text=['caaaar','mussage','survice','hte','u','wtf','g2g']
for word in text:
    print(spell.correction(word))
print(spell.known(text))
```

```python
from autocorrect import Speller

text=['caaaar','mussage','survice','hte','u','wtf','g2g']

spell = Speller(lang='en')

print (spell('caaaar'))
print (spell('mussage'))
print (spell('survice'))
print (spell('hte'))

for word in text:
    print(spell(word))
```
